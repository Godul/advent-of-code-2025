```python
figs = []
problems = []

with open('input.txt', 'r') as file:
    for i in range(6):
        next(file) # number
        figs.append([])
        for i in range(3):
            line = next(file).strip()
            figs[-1].append(line)
        next(file) # endl

    problems = []
    for line in file:
        chunks = line.split()
        x, y = map(int, chunks[0][:-1].split('x'))
        nums = tuple(map(int, chunks[1:]))
        problems.append(((x, y), nums))
```

```python
needed_cells = []

for i, fig in enumerate(figs):
    cells = 0

    for level in fig:
        cells += sum(ch == '#' for ch in level)

    if fig[1][1] != '#':
        cells += 1

    needed_cells.append(cells)
```

```python
maybe_ok = 0
not_ok = 0
ok = 0

for prob in problems:
    (x, y), counts = prob
    needed_min = sum(needed_cells[i] * counts[i] for i in range(len(counts)))
    available = x * y

    needed_naive = sum(counts) * 9

    if available < needed_min:
        not_ok += 1
    elif needed_naive < available or (needed_naive == available and x % 3 == 0 and y % 3 == 0):
        ok += 1
    else:
        maybe_ok += 1

print(f'{not_ok=}, {maybe_ok=}, {ok=}')
```
